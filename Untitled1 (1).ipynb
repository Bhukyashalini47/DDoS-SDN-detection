```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
df1 = pd.read_csv('SDNFlow_Dataset_attack.csv')
df1.head()
```

```python
df2 = pd.read_csv('SDNFlow_Dataset_Normal.csv')
df2.head()
```

```python
df = pd.concat([df1, df2], ignore_index=True)
df.head()
```

```python
df['Category'] = df['Category'].astype('category')
df['flow_ID'] = df['flow_ID'].astype('object')
df['src_port'] = df['src_port'].astype('object')
df['dst_port'] = df['dst_port'].astype('object')
df['TnPPerDport'] = df['TnPPerDport'].astype('object')
```

```python
df.dtypes
```

```python
columns_to_exclude = {'ipv4_src', 'ipv4_dst', 'Category'}
selected_columns = [col for col in df.columns if col not in columns_to_exclude] 
```

```python
scaler = StandardScaler()
df[selected_columns] = scaler.fit_transform(df[selected_columns])

print("Standardized Data:")
print(df)
```

```python
scaler = MinMaxScaler()
df[selected_columns] = pd.DataFrame(scaler.fit_transform(df[selected_columns]), columns= selected_columns)

df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

X = df[selected_columns]  # Features 
y = df['Category']    # Target variable

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize SVM classifier
svm_classifier = SVC(kernel='linear')  # You can choose different kernels depending on your data

# Train SVM classifier
svm_classifier.fit(X_train, y_train)

# Predict on the testing set
y_pred = svm_classifier.predict(X_test)

# Evaluate model performance
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical
from tensorflow.keras import layers, models
import tensorflow as tf  # Add this import statement

# Assuming X_train, X_test, y_train, and y_test are already defined from your SVM code
# Convert X_train and X_test to numpy arrays
X_train_array = np.array(X_train)
X_test_array = np.array(X_test)

# Reshape data for CNN input (example, adjust as needed)
X_train_reshaped = X_train_array.reshape(X_train_array.shape[0], X_train_array.shape[1], 1)
X_test_reshaped = X_test_array.reshape(X_test_array.shape[0], X_test_array.shape[1], 1)

# Encode target variable y to integers
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)

# Calculate the number of classes
num_classes = len(label_encoder.classes_)

# Convert target variable y to one-hot encoded format
y_train_one_hot = to_categorical(y_train_encoded, num_classes=num_classes)
y_test_one_hot = to_categorical(y_test_encoded, num_classes=num_classes)

# Build CNN model with dropout and regularization
model = models.Sequential([
    layers.Conv1D(64, 3, activation='relu', input_shape=(X_train_reshaped.shape[1], 1)),
    layers.MaxPooling1D(2),
    layers.Dropout(0.2),  # Dropout layer
    layers.Flatten(),
    layers.Dense(64, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01)),  # Regularization
    layers.Dense(num_classes, activation='softmax')
])

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model with early stopping
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
model.fit(X_train_reshaped, y_train_one_hot, epochs=50, validation_data=(X_test_reshaped, y_test_one_hot), callbacks=[early_stopping])

# Evaluate model performance
test_loss, test_acc = model.evaluate(X_test_reshaped, y_test_one_hot, verbose=2)
print('Test accuracy:', test_acc)
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming model is your trained CNN model, and X_test_reshaped, y_test_one_hot are your test data
# Predict on the test set
y_pred = model.predict(X_test_reshaped)
y_pred_classes = np.argmax(y_pred, axis=1)

# Convert one-hot encoded labels back to original labels
y_true = np.argmax(y_test_one_hot, axis=1)

# Calculate confusion matrix
conf_matrix = confusion_matrix(y_true, y_pred_classes)

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# Make predictions on the test set
y_pred_probs = model.predict(X_test_reshaped)
y_pred = np.argmax(y_pred_probs, axis=1)

# Convert one-hot encoded y_test back to integer labels for evaluation
y_test_labels = np.argmax(y_test_one_hot, axis=1)

# Calculate precision, recall, and F1 score
precision = precision_score(y_test_labels, y_pred, average='weighted')
recall = recall_score(y_test_labels, y_pred, average='weighted')
f1 = f1_score(y_test_labels, y_pred, average='weighted')

print('Precision:', precision)
print('Recall:', recall)
print('F1 Score:', f1)
```

```python
from sklearn.metrics import precision_recall_curve
import matplotlib.pyplot as plt

# Make predictions on the test set
y_pred_probs = model.predict(X_test_reshaped)

# Calculate precision and recall for each class
precision = dict()
recall = dict()
for i in range(num_classes):  # num_classes is the number of classes in your dataset
    precision[i], recall[i], _ = precision_recall_curve(y_test_one_hot[:, i], y_pred_probs[:, i])

# Plot the precision-recall curves for each class
plt.figure(figsize=(8, 6))
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')

for i in range(num_classes):
    plt.plot(recall[i], precision[i], label=f'Class {i}')

plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# Make predictions on the test set
y_pred_probs = model.predict(X_test_reshaped)

# Calculate false positive rate, true positive rate, and thresholds for each class
fpr = dict()
tpr = dict()
thresholds = dict()
roc_auc = dict()
for i in range(num_classes):  # num_classes is the number of classes in your dataset
    fpr[i], tpr[i], thresholds[i] = roc_curve(y_test_one_hot[:, i], y_pred_probs[:, i])
    roc_auc[i] = roc_auc_score(y_test_one_hot[:, i], y_pred_probs[:, i])

# Plot the ROC curve for each class
plt.figure(figsize=(8, 6))
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')  # Random classifier line
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')

for i in range(num_classes):
    plt.plot(fpr[i], tpr[i], label=f'Class {i} (AUC = {roc_auc[i]:.2f})')

plt.legend()
plt.grid(True)
plt.show()
```
